## üí° User manual :
*Some versions note:*
- Version: All Versions can be used with SD1.5, SDXL only works with ComfyUI, SDXL, A1111-Update, only A1111-Update when running SDXL should have a colabpro account and use the included SDXLMode at part 2 to run with private settings

- Option Prune : Helps to eliminate maximum resources to download, helping to install programs quickly
- Extension Options : Help choosing Extensions to use for Versions A -
- Option Controlnet : Will download the controlnet to use with all versions except SDXL (No controlnet can be used yet), but it will consume a lot of resources, please consider when choosing Extra
- Option Setting : Contains options to update or reset settings, when encountering unexpected errors, try resetting settings
- NoLibrary option : Will remove the download of additional library models, lora is available, if Prune is selected, this option is selected by default
- Swapversion option: Will help to change the Version in use without interrupting the session, if unchecked, rerunning part one only loads more resources that are not available from the previous version

- Option QRControlnet : Only works with Version A - Support downloading more controlnet, extension used during QR art creation
- Option APIPhotoshop : Will help install more Extensions, making the link possible to use API with Photoshop
- Comfy UI : Only works with Prune, Controlnet option
